# **Gene Deletion Simulation**

```python
def reaction_deletion_simulation(reaction_of_interest_id, biomass_val_constraint):
  biomass_dic = {}
  rxn_dic = {}

  if reaction_of_interest_id == 'Biomass': # Handling the case when we want to see the effect of gene deletions on the biomass flux
    model.reactions.get_by_id('Biomass').lower_bound = 0
    model.reactions.get_by_id('Biomass').upper_bound = 1000
    model.objective = 'Biomass'
    # Iterate over reactions
    for reaction in model.reactions:
      # Skip the objective reaction itself
      if (reaction.id == 'Biomass'):
        continue
      # Create a copy of the model to work with
      model_copy = model.copy()

      # Set the knockout constraint for the current reaction in the copy
      model_copy.reactions.get_by_id(reaction.id).lower_bound = 0
      model_copy.reactions.get_by_id(reaction.id).upper_bound = 0

      # Solve the copy of the model to maximize the desired reaction
      solution = model_copy.optimize()

      biomass_dic[reaction.id] = solution.objective_value
    return biomass_dic

  else: # Handling the case when we want to see the effect of gene deletions on the succinic acid flux

    # The goal is to maximize the reaction of interest, given the constraint of the biomass flux as a parameter in the function
    model.reactions.get_by_id('Biomass').lower_bound = biomass_val_constraint
    model.reactions.get_by_id('Biomass').upper_bound = biomass_val_constraint
    model.objective = reaction_of_interest_id # Setting the objective as the reaction of interest after we already constrained the biomass
    for reaction in model.reactions:
        # Skip the biomass reaction (the model has to grow, so biomass rxn cannot be deleted) as well as the reaction of interest
      if (reaction.id == 'Biomass') or (reaction.id == str(reaction_of_interest_id)):
        continue
      # Create a copy of the model to work with so that every time we delete we restore the model again
      model_copy = model.copy()

      # Set the knockout constraint for the current reaction in the copy
      model_copy.reactions.get_by_id(reaction.id).lower_bound = 0
      model_copy.reactions.get_by_id(reaction.id).upper_bound = 0

      # Solve the copy of the model to maximize the desired reaction
      solution = model_copy.optimize()

      # reaction_of_interest = model.reactions.get_by_id(str(reaction_of_interest_id))

      rxn_dic[reaction.id] = solution.objective_value
    return rxn_dic
```

```python
# Extract keys and values
keys = list(reaction_deletion_simulation('Succ_R', 0.3).keys())
values = list(reaction_deletion_simulation('Succ_R', 0.3).values())

plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
plt.bar(keys, values)

plt.xlabel('Reactions')
plt.ylabel('Flux')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

plt.title('Reaction Values')

plt.tight_layout()
plt.show()
```

Based on this plot, it seems that deletion of some reactions are infeasible, and deletion of some other reactions does indeed cause an effect on the succinic acid production.

Particularly, one reaction is unique in that its deletion causes the most increase in the flux of succinic acid exchange. This reaction is the **RPI** reaction where the flux is increased to 1.4. This is 16% increase.
